O conjunto de dados Linnerud é um pequeno conjunto clássico usado em Estatística e Aprendizado de Máquina, composto por 20 observações que relacionam 3 variáveis de exercício físico (Chins - flexões, Situps - abdominais, Jumps - saltos) com 3 variáveis fisiológicas (Weight - peso, Waist - cintura, Pulse - frequência cardíaca)

```python
from sklearn.datasets import load_linnerud
from sklearn.datasets import load_linnerud
import pandas as pd

# carregando o conjunto de dados
linnerud = load_linnerud()

# convertendo para DataFrame
df = pd.DataFrame(linnerud.data, columns=linnerud.feature_names)
df_target = pd.DataFrame(linnerud.target, columns=linnerud.target_names)

# juntando eles
df = pd.concat([df, df_target], axis=1)

# printar
df.head()
```

```python
def Media(Weight):
    return df[Weight].mean()

def DesvioPadrao(Weight):
    return df[Weight].std()

def IntervaloInterquartil(Jumps):
    return df[Jumps].quantile(0.75) - df[Jumps].quantile(0.25)

print(f"A média dos pesos dos candidatos é: {Media('Weight')}")
print(f"O desvio padrão dos pesos dos candidatos é: {DesvioPadrao('Weight')}")
print(f"O intervalo Interquartil dos números de saltos dos candidatos é: {IntervaloInterquartil('Jumps')}")
```

```python
import matplotlib.pyplot as plt

chins = df['Chins'].tolist()
pulse = df['Pulse'].tolist()

plt.scatter(chins, pulse)
plt.xlabel("Flexões")
plt.ylabel("Pulso")
plt.show()
```
